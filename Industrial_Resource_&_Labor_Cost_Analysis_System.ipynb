```python
import pandas as pd

industry_knowledge = {
    "Steel": {
        "Type": "Heavy Industry",
        "Land": "Flat, Load-Bearing",
        "Water": "High (Large River)",
        "Daily_Wage": 50,
        "Staff": 200,
        "Description": "Production of alloys from iron ore and coal."
    },
    "Silk": {
        "Type": "Agro-Textile",
        "Land": "Fertile, Tropical",
        "Water": "Medium (Streams)",
        "Daily_Wage": 15,
        "Staff": 500,
        "Description": "Rearing silkworms and reeling natural fibers."
    },
    "Glass": {
        "Type": "Manufacturing",
        "Land": "Industrial Zone",
        "Water": "Low",
        "Daily_Wage": 40,
        "Staff": 100,
        "Description": "Melting silica sand into transparent sheets/containers."
    },
    "Cement": {
        "Type": "Construction",
        "Land": "Near Quarries",
        "Water": "Medium",
        "Daily_Wage": 30,
        "Staff": 150,
        "Description": "Processing limestone for building materials."
    },
    "Software": {
        "Type": "Technology",
        "Land": "Urban Tech Park",
        "Water": "Minimal",
        "Daily_Wage": 120,
        "Staff": 50,
        "Description": "Digital services and code development."
    }
}



def display_menu():
    print("\n" + "="*40)
    print("INDUSTRIAL KNOWLEDGE SYSTEM")
    print("="*40)
    print("Our database currently contains info on:")
    for key in industry_knowledge.keys():
        print(f" - {key}")
    print("="*40)

def run_analysis(user_input):
    name = user_input.capitalize().strip()


    if name in industry_knowledge:
        data = industry_knowledge[name]
        monthly_labor = data['Daily_Wage'] * data['Staff'] * 22

        print(f"\n DATA FOUND FOR: {name.upper()}")
        print(f"Sector: {data['Type']}")
        print(f"About: {data['Description']}")
        print(f"Required Land: {data['Land']}")
        print(f"Water Need: {data['Water']}")
        print(f"Staff Count: {data['Staff']}")
        print(f"Monthly Labor Cost: ${monthly_labor:,}")
    else:
        print(f"\n ERROR: '{user_input}' is not in our database.")
        print("Please choose from the list provided in the menu.")
display_menu()

while True:
    choice = input("\nEnter an industry from the list (or 'exit'): ")
    if choice.lower() == 'exit':
        print("Goodbye!")
        break
    run_analysis(choice)
```
